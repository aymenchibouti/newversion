```python
# -*- coding: utf-8 -*-
"""
Enhanced Deep Learning Pipeline for Dropout Prediction
- Uses TensorFlow/Keras with advanced architectures
- Implements multiple model architectures (Dense, ResNet-style, TabNet-inspired)
- Advanced preprocessing and feature engineering
- Comprehensive evaluation and visualization
- SHAP integration for deep learning models
"""

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_auc_score, roc_curve
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline as ImbPipeline

# Deep Learning imports
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, regularizers, callbacks
from tensorflow.keras.optimizers import Adam, AdamW
from tensorflow.keras.losses import BinaryCrossentropy
from tensorflow.keras.metrics import AUC, Precision, Recall

# Explainability
import shap

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Configuration
CONFIG = {
    'file_path': 'model1_210_features.csv',
    'random_state': 42,
    'test_size': 0.2,
    'validation_size': 0.2,
    'batch_size': 512,
    'epochs': 100,
    'early_stopping_patience': 15,
    'reduce_lr_patience': 8,
    'learning_rate': 0.001
}

class DeepLearningDropoutPredictor:
    """
    A comprehensive deep learning system for student dropout prediction
    """

    def __init__(self, config):
        self.config = config
        self.scaler = None
        self.models = {}
        self.histories = {}
        self.feature_names = None

    def load_and_preprocess_data(self):
        """Load and preprocess the dataset"""
        print("Loading and preprocessing data...")

        # Load data
        if not os.path.exists(self.config['file_path']):
            raise FileNotFoundError(f"Could not find '{self.config['file_path']}'")

        data = pd.read_csv(self.config['file_path'])
        print(f"Dataset shape: {data.shape}")

        # Remove identifier columns
        id_columns = ['username', 'course_id', 'enrollment_id']
        data = data.drop(columns=[col for col in id_columns if col in data.columns])

        # Handle missing values
        data = data.fillna(0)

        # Ensure dropout column exists and is binary
        if 'dropout' not in data.columns:
            raise KeyError("Column 'dropout' not found in the dataset")

        # Separate features and target
        X = data.drop(columns=['dropout'])
        y = data['dropout']

        self.feature_names = X.columns.tolist()
        print(f"Number of features: {len(self.feature_names)}")
        print(f"Class distribution: {y.value_counts().to_dict()}")

        return X, y

    def advanced_feature_engineering(self, X):
        """Apply advanced feature engineering techniques"""
        print("Applying feature engineering...")

        X_engineered = X.copy()

        # Create daily activity aggregations
        days = range(1, 31)  # 30 days
        activities = ['access', 'problem', 'wiki', 'discussion', 'navigate', 'page_close', 'video']

        # Weekly aggregations
        for week in range(1, 5):  # 4 weeks
            week_start = (week - 1) * 7 + 1
            week_end = min(week * 7, 30)

            for activity in activities:
                week_cols = [f'day_{day}_{activity}' for day in range(week_start, week_end + 1)
                           if f'day_{day}_{activity}' in X.columns]
                if week_cols:
                    X_engineered[f'week_{week}_{activity}_total'] = X[week_cols].sum(axis=1)
                    X_engineered[f'week_{week}_{activity}_mean'] = X[week_cols].mean(axis=1)
                    X_engineered[f'week_{week}_{activity}_max'] = X[week_cols].max(axis=1)

        # Overall activity patterns
        for activity in activities:
            activity_cols = [col for col in X.columns if activity in col]
            if activity_cols:
                X_engineered[f'total_{activity}'] = X[activity_cols].sum(axis=1)
                X_engineered[f'mean_{activity}'] = X[activity_cols].mean(axis=1)
                X_engineered[f'std_{activity}'] = X[activity_cols].std(axis=1).fillna(0)
                X_engineered[f'max_{activity}'] = X[activity_cols].max(axis=1)
                # Activity consistency (coefficient of variation)
                mean_activity = X[activity_cols].mean(axis=1)
                std_activity = X[activity_cols].std(axis=1).fillna(0)
                X_engineered[f'cv_{activity}'] = np.where(mean_activity != 0,
                                                        std_activity / mean_activity, 0)

        # Engagement trends
        X_engineered['early_engagement'] = X[[f'day_{i}_access' for i in range(1, 8)
                                            if f'day_{i}_access' in X.columns]].sum(axis=1)
        X_engineered['late_engagement'] = X[[f'day_{i}_access' for i in range(24, 31)
                                           if f'day_{i}_access' in X.columns]].sum(axis=1)
        X_engineered['engagement_decline'] = X_engineered['early_engagement'] - X_engineered['late_engagement']

        # Active days count
        access_cols = [col for col in X.columns if 'access' in col]
        if access_cols:
            X_engineered['active_days'] = (X[access_cols] > 0).sum(axis=1)
            X_engineered['activity_ratio'] = X_engineered['active_days'] / len(access_cols)

        print(f"Features after engineering: {X_engineered.shape[1]}")
        return X_engineered

    def prepare_data(self, X, y):
        """Prepare data for training with advanced preprocessing"""
        print("Preparing data for training...")

        # Apply feature engineering
        X_engineered = self.advanced_feature_engineering(X)

        # Split data
        X_temp, X_test, y_temp, y_test = train_test_split(
            X_engineered, y, test_size=self.config['test_size'],
            random_state=self.config['random_state'], stratify=y
        )

        X_train, X_val, y_train, y_val = train_test_split(
            X_temp, y_temp, test_size=self.config['validation_size'],
            random_state=self.config['random_state'], stratify=y_temp
        )

        # Scale features using RobustScaler (less sensitive to outliers)
        self.scaler = RobustScaler()
        X_train_scaled = self.scaler.fit_transform(X_train)
        X_val_scaled = self.scaler.transform(X_val)
        X_test_scaled = self.scaler.transform(X_test)

        # Handle class imbalance with SMOTE
        smote = SMOTE(sampling_strategy='auto', random_state=self.config['random_state'])
        X_train_balanced, y_train_balanced = smote.fit_resample(X_train_scaled, y_train)

        print(f"Training set shape after balancing: {X_train_balanced.shape}")
        print(f"Validation set shape: {X_val_scaled.shape}")
        print(f"Test set shape: {X_test_scaled.shape}")

        # Update feature names
        self.feature_names = X_engineered.columns.tolist()

        return (X_train_balanced, y_train_balanced, X_val_scaled, y_val,
                X_test_scaled, y_test, X_train_scaled, y_train)

    def create_dense_model(self, input_dim):
        """Create a deep dense neural network"""
        model = keras.Sequential([
            layers.Input(shape=(input_dim,)),

            # First block
            layers.Dense(512, activation='relu'),
            layers.BatchNormalization(),
            layers.Dropout(0.3),

            # Second block
            layers.Dense(256, activation='relu'),
            layers.BatchNormalization(),
            layers.Dropout(0.3),

            # Third block
            layers.Dense(128, activation='relu'),
            layers.BatchNormalization(),
            layers.Dropout(0.2),

            # Fourth block
            layers.Dense(64, activation='relu'),
            layers.BatchNormalization(),
            layers.Dropout(0.2),

            # Output
            layers.Dense(1, activation='sigmoid')
        ])

        return model

    def create_resnet_style_model(self, input_dim):
        """Create a ResNet-style model with skip connections"""
        inputs = layers.Input(shape=(input_dim,))

        # Initial dense layer
        x = layers.Dense(512, activation='relu')(inputs)
        x = layers.BatchNormalization()(x)

        # ResNet block 1
        residual = x
        x = layers.Dense(512, activation='relu')(x)
        x = layers.BatchNormalization()(x)
        x = layers.Dropout(0.3)(x)
        x = layers.Dense(512, activation='relu')(x)
        x = layers.BatchNormalization()(x)
        x = layers.Add()([x, residual])
        x = layers.Activation('relu')(x)
        x = layers.Dropout(0.3)(x)

        # Transition layer
        x = layers.Dense(256, activation='relu')(x)
        x = layers.BatchNormalization()(x)

        # ResNet block 2
        residual = x
        x = layers.Dense(256, activation='relu')(x)
        x = layers.BatchNormalization()(x)
        x = layers.Dropout(0.2)(x)
        x = layers.Dense(256, activation='relu')(x)
        x = layers.BatchNormalization()(x)
        x = layers.Add()([x, residual])
        x = layers.Activation('relu')(x)
        x = layers.Dropout(0.2)(x)

        # Final layers
        x = layers.Dense(128, activation='relu')(x)
        x = layers.BatchNormalization()(x)
        x = layers.Dropout(0.1)(x)

        outputs = layers.Dense(1, activation='sigmoid')(x)

        model = keras.Model(inputs, outputs)
        return model

    def create_attention_model(self, input_dim):
        """Create a model with attention mechanism for tabular data"""
        inputs = layers.Input(shape=(input_dim,))

        # Feature embedding
        x = layers.Dense(256, activation='relu')(inputs)
        x = layers.BatchNormalization()(x)
        x = layers.Dropout(0.3)(x)

        # Attention mechanism
        # Query, Key, Value projections
        query = layers.Dense(128)(x)
        key = layers.Dense(128)(x)
        value = layers.Dense(128)(x)

        # Attention weights
        attention_weights = layers.Dot(axes=1)([query, key])
        attention_weights = layers.Activation('softmax')(attention_weights)

        # Apply attention
        attended = layers.Multiply()([attention_weights, value])

        # Combine with original features
        combined = layers.Concatenate()([x, attended])

        # Final layers
        x = layers.Dense(128, activation='relu')(combined)
        x = layers.BatchNormalization()(x)
        x = layers.Dropout(0.2)(x)

        x = layers.Dense(64, activation='relu')(x)
        x = layers.BatchNormalization()(x)
        x = layers.Dropout(0.1)(x)

        outputs = layers.Dense(1, activation='sigmoid')(x)

        model = keras.Model(inputs, outputs)
        return model

    def compile_model(self, model, name):
        """Compile model with appropriate optimizer and metrics"""
        # Use AdamW optimizer for better generalization
        optimizer = AdamW(learning_rate=self.config['learning_rate'], weight_decay=1e-5)

        model.compile(
            optimizer=optimizer,
            loss=BinaryCrossentropy(),
            metrics=[
                'accuracy',
                AUC(name='auc'),
                Precision(name='precision'),
                Recall(name='recall')
            ]
        )

        print(f"\n{name} Model Architecture:")
        model.summary()
        return model

    def get_callbacks(self, model_name):
        """Get training callbacks"""
        callbacks_list = [
            callbacks.EarlyStopping(
                monitor='val_auc',
                patience=self.config['early_stopping_patience'],
                restore_best_weights=True,
                mode='max',
                verbose=1
            ),
            callbacks.ReduceLROnPlateau(
                monitor='val_auc',
                factor=0.5,
                patience=self.config['reduce_lr_patience'],
                min_lr=1e-7,
                mode='max',
                verbose=1
            ),
            callbacks.ModelCheckpoint(
                filepath=f'best_{model_name}_model.h5',
                monitor='val_auc',
                save_best_only=True,
                mode='max',
                verbose=1
            )
        ]

        return callbacks_list

    def train_models(self, X_train, y_train, X_val, y_val):
        """Train multiple model architectures"""
        input_dim = X_train.shape[1]

        # Model configurations
        model_configs = {
            'Dense': self.create_dense_model,
            'ResNet': self.create_resnet_style_model,
            'Attention': self.create_attention_model
        }

        for name, create_func in model_configs.items():
            print(f"\n{'='*50}")
            print(f"Training {name} Model")
            print(f"{'='*50}")

            # Create and compile model
            model = create_func(input_dim)
            model = self.compile_model(model, name)

            # Train model
            history = model.fit(
                X_train, y_train,
                batch_size=self.config['batch_size'],
                epochs=self.config['epochs'],
                validation_data=(X_val, y_val),
                callbacks=self.get_callbacks(name.lower()),
                verbose=1
            )

            # Store model and history
            self.models[name] = model
            self.histories[name] = history

        print("\nAll models trained successfully!")

    def evaluate_models(self, X_test, y_test):
        """Evaluate all trained models"""
        results = {}

        print("\n" + "="*60)
        print("MODEL EVALUATION RESULTS")
        print("="*60)

        for name, model in self.models.items():
            # Predictions
            y_pred_proba = model.predict(X_test, verbose=0)
            y_pred = (y_pred_proba > 0.5).astype(int)

            # Calculate metrics
            accuracy = accuracy_score(y_test, y_pred)
            precision = precision_score(y_test, y_pred, zero_division=0)
            recall = recall_score(y_test, y_pred, zero_division=0)
            f1 = f1_score(y_test, y_pred, zero_division=0)
            auc = roc_auc_score(y_test, y_pred_proba)

            results[name] = {
                'accuracy': accuracy,
                'precision': precision,
                'recall': recall,
                'f1_score': f1,
                'auc': auc,
                'predictions': y_pred,
                'probabilities': y_pred_proba
            }

            print(f"\n{name} Model Results:")
            print(f"  Accuracy:  {accuracy:.4f}")
            print(f"  Precision: {precision:.4f}")
            print(f"  Recall:    {recall:.4f}")
            print(f"  F1 Score:  {f1:.4f}")
            print(f"  AUC:       {auc:.4f}")

        return results

    def plot_training_history(self):
        """Plot training history for all models"""
        fig, axes = plt.subplots(2, 2, figsize=(15, 12))
        fig.suptitle('Training History Comparison', fontsize=16)

        metrics = ['accuracy', 'auc', 'loss', 'precision']

        for i, metric in enumerate(metrics):
            ax = axes[i//2, i%2]

            for name, history in self.histories.items():
                if metric in history.history:
                    ax.plot(history.history[metric], label=f'{name} Train', linestyle='-')
                    ax.plot(history.history[f'val_{metric}'], label=f'{name} Val', linestyle='--')

            ax.set_title(f'{metric.capitalize()} History')
            ax.set_xlabel('Epochs')
            ax.set_ylabel(metric.capitalize())
            ax.legend()
            ax.grid(True, alpha=0.3)

        plt.tight_layout()
        plt.show()

    def plot_model_comparison(self, results):
        """Plot comparison of model performance"""
        metrics = ['accuracy', 'precision', 'recall', 'f1_score', 'auc']
        model_names = list(results.keys())

        fig, ax = plt.subplots(figsize=(12, 8))

        x = np.arange(len(metrics))
        width = 0.25

        for i, model in enumerate(model_names):
            values = [results[model][metric] for metric in metrics]
            ax.bar(x + i*width, values, width, label=model, alpha=0.8)

        ax.set_xlabel('Metrics')
        ax.set_ylabel('Score')
        ax.set_title('Model Performance Comparison')
        ax.set_xticks(x + width)
        ax.set_xticklabels([m.replace('_', ' ').title() for m in metrics])
        ax.legend()
        ax.grid(True, alpha=0.3)

        # Add value labels on bars
        for i, model in enumerate(model_names):
            values = [results[model][metric] for metric in metrics]
            for j, value in enumerate(values):
                ax.text(j + i*width, value + 0.01, f'{value:.3f}',
                       ha='center', va='bottom', fontsize=9)

        plt.tight_layout()
        plt.show()

    def plot_confusion_matrices(self, results, y_test):
        """Plot confusion matrices for all models"""
        n_models = len(results)
        fig, axes = plt.subplots(1, n_models, figsize=(5*n_models, 4))

        if n_models == 1:
            axes = [axes]

        for i, (name, result) in enumerate(results.items()):
            cm = confusion_matrix(y_test, result['predictions'])

            sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[i])
            axes[i].set_title(f'{name} Model\nConfusion Matrix')
            axes[i].set_xlabel('Predicted')
            axes[i].set_ylabel('Actual')

        plt.tight_layout()
        plt.show()

    def plot_roc_curves(self, results, y_test):
        """Plot ROC curves for all models"""
        plt.figure(figsize=(10, 8))

        for name, result in results.items():
            fpr, tpr, _ = roc_curve(y_test, result['probabilities'])
            auc_score = result['auc']
            plt.plot(fpr, tpr, label=f'{name} (AUC = {auc_score:.3f})', linewidth=2)

        plt.plot([0, 1], [0, 1], 'k--', label='Random Classifier')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('ROC Curves Comparison')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()

    def explain_best_model(self, X_test, y_test, X_train_original):
        """Explain the best performing model using SHAP"""
        # Find best model by AUC
        best_model_name = max(self.models.keys(),
                            key=lambda k: roc_auc_score(y_test, self.models[k].predict(X_test)))
        best_model = self.models[best_model_name]

        print(f"\nGenerating SHAP explanations for best model: {best_model_name}")

        try:
            # Create SHAP explainer for deep learning model
            explainer = shap.DeepExplainer(best_model, X_train_original[:100])  # Use subset for efficiency
            shap_values = explainer.shap_values(X_test[:100])  # Explain first 100 test samples

            # Summary plot
            plt.figure(figsize=(10, 8))
            shap.summary_plot(shap_values, X_test[:100], feature_names=self.feature_names, show=False)
            plt.title(f'SHAP Summary Plot - {best_model_name} Model')
            plt.tight_layout()
            plt.show()

            # Feature importance plot
            plt.figure(figsize=(10, 8))
            shap.summary_plot(shap_values, X_test[:100], feature_names=self.feature_names,
                            plot_type="bar", show=False)
            plt.title(f'SHAP Feature Importance - {best_model_name} Model')
            plt.tight_layout()
            plt.show()

        except Exception as e:
            print(f"SHAP explanation failed: {str(e)}")
            print("This might be due to memory constraints or model complexity.")

    def run_complete_pipeline(self):
        """Run the complete deep learning pipeline"""
        print("Starting Deep Learning Pipeline for Dropout Prediction")
        print("="*60)

        # Load and preprocess data
        X, y = self.load_and_preprocess_data()

        # Prepare data for training
        (X_train, y_train, X_val, y_val,
         X_test, y_test, X_train_original, y_train_original) = self.prepare_data(X, y)

        # Train models
        self.train_models(X_train, y_train, X_val, y_val)

        # Evaluate models
        results = self.evaluate_models(X_test, y_test)

        # Create visualizations
        self.plot_training_history()
        self.plot_model_comparison(results)
        self.plot_confusion_matrices(results, y_test)
        self.plot_roc_curves(results, y_test)

        # Explain best model
        self.explain_best_model(X_test, y_test, X_train_original)

        print("\n" + "="*60)
        print("Deep Learning Pipeline Completed Successfully!")
        print("="*60)

        return results

# Main execution
def main():
    """Main function to run the deep learning pipeline"""
    # Check TensorFlow GPU availability
    print("TensorFlow version:", tf.__version__)
    print("GPU Available:", tf.config.list_physical_devices('GPU'))

    # Create and run the predictor
    predictor = DeepLearningDropoutPredictor(CONFIG)
    results = predictor.run_complete_pipeline()

    # Print final summary
    print("\nFinal Model Rankings (by AUC):")
    rankings = sorted(results.items(), key=lambda x: x[1]['auc'], reverse=True)
    for i, (name, metrics) in enumerate(rankings, 1):
        print(f"{i}. {name}: AUC = {metrics['auc']:.4f}, Accuracy = {metrics['accuracy']:.4f}")

if __name__ == '__main__':
    main()
```
